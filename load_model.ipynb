```python
# Loading the model
# WARNING: Never try to print the represention of the model -> it will freeze due to its size
model = Doc2Vec.load("modelv2.doc2vec")
```

```python
def repr_doc2vec():
    """
    Represensation of instances as a vector with n entries
    """
    print("single vector 9135 : ", model.dv[9135])
    print('--------------------------------------------------------------------------')
    # Access specific entry of a vector of an instance
    print("single feature of instance 9135 : ", model.dv[9135][0])
    print('--------------------------------------------------------------------------')
    # Access specific range of entries of a vector of an instance
    print("single vector with range 0:5: ", model.dv[9135][0:5])
    
    
repr_doc2vec()
```
